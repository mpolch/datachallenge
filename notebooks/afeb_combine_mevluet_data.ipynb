```python
# Read Mevlet data file

from zipfile import ZipFile
import pandas as pd
```

```python
with ZipFile('../collection_csv.zip') as myzip:
    with myzip.open('collection_csv/us-balance-annual.csv') as f:
        df_balance = pd.read_csv(f, sep=';')
    with myzip.open('collection_csv/us-income-annual.csv') as f:
        df_income = pd.read_csv(f, sep=';')
    with myzip.open('collection_csv/us-companies.csv') as f:
        df_companies = pd.read_csv(f, sep=';')
    with myzip.open('collection_csv/us-cashflow-annual.csv') as f:
        df_cashflow = pd.read_csv(f, sep=';')
    with myzip.open('collection_csv/industries.csv') as f:
        df_industry = pd.read_csv(f, sep=';')
    with myzip.open('collection_csv/industries.csv') as f:
        df_industry = pd.read_csv(f, sep=';')
    with myzip.open('collection_csv/markets.csv') as f:
        df_market = pd.read_csv(f, sep=';')
```

```python
# upper-case company name

df_companies['Company Name'] = df_companies['Company Name'].str.upper()
```

```python
# merge income with companies to have company name

df_income_merge = pd.merge(df_income, df_companies, on='Ticker', how='inner', validate='many_to_one')
df_income_merge['Company Name']
```

```python
# read list of companies that survied and died the 2008 crisis

df_survived = pd.read_csv("../data/processed/companies_survived.csv", index_col=0)
df_died = pd.read_csv("../data/processed/companies_died.csv", index_col=0)

print(df_survived.shape, df_died.shape)
```

```python
# merge income dataset with companies that survived on company name

df_survived_income = pd.merge(
    df_survived, df_income_merge, on="Company Name", validate="one_to_many"
)
df_died_income = pd.merge(
    df_died, df_income_merge, on="Company Name", validate="one_to_many"
)
```
