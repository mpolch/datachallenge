### Healthcare Sector

```python
df_h = df_m[df_m['Sector'] == "Healthcare"]
```

```python
df_liquid = df_h["Total Current Assets"]/df_h["Total Current Liabilities"]
df_h["liquid"] = df_liquid
```

```python
xx = "Fiscal Year"
```

```python
yy = "Total Current Assets"
```

```python
df_h.plot.scatter(xx, yy) # Fiscal Year vs 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(data=df_h, x=xx, y="liquid", ax=ax)
ax.set_xlabel("Fiscal Year", size = 18)
ax.set_ylabel('Liquidity', size = 18)
ax.set_title('Liquidity in Healthcare Sector in Years', size = 20)
```

```python
df_solvency = df_h["Long Term Debt"]/df_h["Total Equity"]
df_h["solvency"] = df_solvency
```
